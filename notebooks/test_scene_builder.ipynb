```python
import os, psutil

def print_memory_usage():
    process = psutil.Process(os.getpid())
    memory_info = process.memory_info()
    print(f"Memory usage: {memory_info.rss / 1024 ** 2:.2f} MB")

print_memory_usage()
# log_names = ["2021.06.07.12.54.00_veh-35_01843_02314"]
# log_names = ["_Rep0_bench2drive220_route2_06_12_20_50_31"]
log_names = None
split = "carla"
scene_filter = SceneFilter(split_names=[split], log_names=log_names, timestamp_threshold_s=60.0, duration_s=60.0)
scene_builder = ArrowSceneBuilder("/home/daniel/asim_workspace/data")
worker = Sequential()  
# worker = RayDistributed()
scenes = scene_builder.get_scenes(scene_filter, worker)

print(len(scenes))
print_memory_usage()
```

```python
from pathlib import Path
from asim.common.visualization.matplotlib.plots import render_scene_animation

output_path = Path("/home/daniel/asim_route_testing")
for scene in scenes:
    render_scene_animation(scene, output_path, fps=30, end_idx=None, step=5)
```
