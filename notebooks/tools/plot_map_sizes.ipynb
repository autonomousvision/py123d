```python
import os

import matplotlib.pyplot as plt

folder_path = "/home/daniel/asim_workspace/data/maps"  # Replace with your folder path

# Get all files in the folder (not subfolders)
files = sorted([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
sizes = [os.path.getsize(os.path.join(folder_path, f)) / 1024**2 for f in files]

plt.figure(figsize=(10, 6))
plt.bar(files, sizes)
plt.ylabel("File Size (MB)")
plt.title("Map Sizes (CARLA/nuPlan)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
