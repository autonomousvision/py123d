```python
from pathlib import Path
```

```python
import gzip
import json
from typing import Dict


path = Path("/home/daniel/carla_workspace/data/_Rep0_longest1_route0_07_04_10_18_47/boxes")

def _load_json_gz(path: Path) -> Dict:
    """Helper function to load a gzipped JSON file."""
    with gzip.open(path, "rt") as f:
        data = json.load(f)
    return data

bounding_box_paths = sorted([bb_path for bb_path in path.iterdir()])
```

```python

width = [] 
length = [] 
height = []
wheel_base = [] 
rear_axle_to_center_vertical = [] 
rear_axle_to_center_longitudinal = [] 
for bb_path in bounding_box_paths:
    bb_dict = _load_json_gz(bb_path)
    vehicle_parameters = bb_dict["vehicle_parameters"]
    width.append(vehicle_parameters["width"])
    length.append(vehicle_parameters["length"])
    height.append(vehicle_parameters["height"])
    wheel_base.append(vehicle_parameters["wheel_base"])
    rear_axle_to_center_vertical.append(vehicle_parameters["rear_axle_to_center_vertical"])
    rear_axle_to_center_longitudinal.append(vehicle_parameters["rear_axle_to_center_longitudinal"])
```

```python
import numpy as np

print("width", np.mean(width), np.std(width))
print("length", np.mean(length), np.std(length))
print("height", np.mean(height), np.std(height))
print("wheel_base", np.mean(wheel_base), np.std(wheel_base))
print("rear_axle_to_center_vertical", np.mean(rear_axle_to_center_vertical), np.std(rear_axle_to_center_vertical))
print("rear_axle_to_center_longitudinal", np.mean(rear_axle_to_center_longitudinal), np.std(rear_axle_to_center_longitudinal))


```

```python
print("width", round(np.median(width), 5), np.std(width))
print("length", round(np.median(length), 5), np.std(length))
print("height", round(np.median(height), 5), np.std(height))
print("wheel_base", round(np.median(wheel_base), 5), np.std(wheel_base))
print("rear_axle_to_center_vertical", round(np.median(rear_axle_to_center_vertical), 5), np.std(rear_axle_to_center_vertical))
print("rear_axle_to_center_longitudinal", round(np.median(rear_axle_to_center_longitudinal), 5), np.std(rear_axle_to_center_longitudinal))
```
